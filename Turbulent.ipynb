```python
import numpy as np
```

```python
arr = np.random.uniform(low=-1.0,high=1.0,size=(10,10))
```

```python
print(arr)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(arr);
```

```python
from matplotlib import pyplot
```

```python
pyplot.figure(figsize=(7,7))
pyplot.imshow(arr)
pyplot.colorbar()
pyplot.show()
```

```python
import scipy.linalg as la
```

```python
eigvals, eigvecs = la.eig(arr)
print(eigvals)
```

```python
print(eigvecs)
```

```python
# Applying Schmidt decomposition on the matrix

la.svd(arr, full_matrices=True)
```

```python
la.svdvals(arr)
```

```python
eigvec = np.real(eigvecs)
eigvec
```

```python
eigvec = np.array(eigvec)
eigvec
```

```python
eigvec_Dirty = eigvec+0.00001*np.random.rand(10, 10)
```

```python
eigvec_inv = la.inv(eigvec_Dirty)
eigvec_inv
```

```python
D = eigvec_inv.dot(arr).dot(eigvec_Dirty)
D
```

```python
A = eigvec_Dirty.dot(D).dot(eigvec_inv)
A
```

```python
arr
```

```python
plt.plot(A);
```

```python
pyplot.figure(figsize=(7,7))
pyplot.imshow(A)
pyplot.colorbar()
pyplot.show()
```

```python
fig, ax = plt.subplots()
ax.hist(A);
```
